```python
# Calculating Defect Removal Effectiveness for release 1

sumcol = release_1.sum(axis=0)

sumrow = release_1.sum(axis=1)

for i in range(0, len(sumrow)):
    release_1.loc[i, 'Detected'] = sumrow[i]

for i in range(0, len(sumrow)):
    release_1.loc[i, 'Injected'] = sumcol[i+1]

release_1.loc[0, 'Escaped'] = release_1.loc[0, 'Injected'] - release_1.loc[0, 'Detected']
for i in range(1, len(sumrow)):
    release_1.loc[i, 'Escaped'] = (release_1.loc[i, 'Injected'] - release_1.loc[i, 'Detected']) + release_1.loc[i-1, 'Escaped']
for i in range(1, len(sumrow)):
    release_1.loc[i, 'DefectsOnStepEntry'] = release_1.loc[i-1, 'Escaped']

# calculating DRE for inspection phase
release_1.loc[0, 'Defect Removal Effectiveness'] = (release_1.loc[0, 'Detected'] / release_1.loc[0, 'Injected']) * 100
for i in range(1, 4):
    detected_sum = 0
    for j in range(0, i):
        detected_sum += release_1.loc[j, 'Detected']
    injected_sum = 0
    for j in range(0, i+1):
        injected_sum += release_1.loc[j, 'Injected']
    release_1.loc[i, 'Defect Removal Effectiveness'] = (release_1.loc[i, 'Detected'] / (injected_sum - detected_sum)) * 100
# calculating DRE for testing phase
for i in range(4,7):
    detected_sum = 0
    for j in range(i, 8):
        detected_sum += release_1.loc[j, 'Detected']
    release_1.loc[i, 'Defect Removal Effectiveness'] = (release_1.loc[i, 'Detected'] / detected_sum) * 100
print(release_1)
```
